```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data=pd.read_csv('/content/sample_data/Dataset1 .csv')
```

```python
data['Restaurant Name Length'] = data['Restaurant Name'].apply(len)
data['Address Length'] = data['Address'].apply(len)

print("Sample data with new length features:")
print(data[['Restaurant Name', 'Restaurant Name Length', 'Address', 'Address Length']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.histplot(data['Restaurant Name Length'], kde=True, color='GREEN')
plt.title('Distribution of Restaurant Name Length')
plt.xlabel('Length of Restaurant Name')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data['Address Length'], kde=True, color='BLUE')
plt.title('Distribution of Address Length')
plt.xlabel('Length of Address')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Has Table Booking (Encoded)'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})

data['Has Online Delivery (Encoded)'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})
print("nSample data with new encoded features:")
print(data[['Has Table booking', 'Has Table Booking (Encoded)', 'Has Online delivery', 'Has Online Delivery (Encoded)']].head())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Has Online Delivery (Encoded)', data=data, palette='hot')
plt.title('Distribution of Restaurants with Online Delivery')
plt.xlabel('Has Online Delivery (Encoded)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='Has Table Booking (Encoded)', data=data, palette='Set2')
plt.title('Distribution of Restaurants with Table Booking')
plt.xlabel('Has Table Booking (Encoded)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
